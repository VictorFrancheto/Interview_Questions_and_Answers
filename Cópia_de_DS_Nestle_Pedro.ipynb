```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2024)
# 1. Carregar o conjunto de dados em um DataFrame
num_rows = 100

# Definir a lista de produtos
produtos = ['Produto A', 'Produto B', 'Produto C']

# Inicializar listas vazias para armazenar os dados
data = {
    'Data': [],
    'Produto': [],
    'Quantidade': [],
    'Preço': []
}

# Simular os dados
for _ in range(num_rows):
    data['Data'].append(np.random.choice(pd.date_range(start='2022-01-01', end='2022-12-31')))
    data['Produto'].append(np.random.choice(produtos))
    data['Quantidade'].append(np.random.randint(1, 10))
    data['Preço'].append(np.round(np.random.uniform(10, 100), 2))

# Criar o DataFrame
df = pd.DataFrame(data)
df.head()
```
